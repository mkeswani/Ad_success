# Ad-success Exploratory Data Analysis 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
ad_data = pd.read_csv('/Users/maheema/Desktop/Data Science - Personal /Ad_Success/Data/Train.csv')
```

**What percentage of ads were successful?**

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
ad_data['netgain'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('netgain')
ax[0].set_ylabel('')
sns.countplot('netgain',data=ad_data,ax=ax[1])
ax[1].set_title('netgain')
plt.show()
```

From the graph above, we see only 1 in 4 ads are successful, on average. To get a better understanding of which factors are likely more significant, we need to do some exploratory data work and see the distributions of other variables in the data set.
